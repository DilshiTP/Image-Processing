```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
f = cv.imread(r'C:\Users\Asus\Desktop\Image Processing\spider.png', cv.IMREAD_GRAYSCALE)
assert f is not None

gamma = 2
t = np.array([(p/255)**gamma*255 for p in range(0,256)]).astype(np.uint8)        #gamma correction
g = cv.LUT(f,t)

fig,ax = plt.subplots()
ax.plot(t)

cv.namedWindow('Image',cv.WINDOW_AUTOSIZE)
cv.imshow('Image',f)
cv.waitKey(0)
cv.imshow('Image',g)
cv.waitKey(0)
cv.destroyAllWindows()

h = cv.cvtColor(f,cv.COLOR_BGR2RGB)
i = cv.cvtColor(g,cv.COLOR_BGR2RGB)
fig,ax = plt.subplots(2,1,figsize = (17,10))
ax[0].imshow(h)
ax[1].imshow(i)
plt.show
```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

f = cv.imread(r'C:\Users\Asus\Desktop\Image Processing\spider.png', cv.IMREAD_GRAYSCALE)
assert f is not None

t1 = np.linspace(0,100,100)
t2 = np.linspace(100,255,50)
t3 = np.linspace(255,255,106)
t = np.concatenate((t1,t2,t3),axis=0).astype(np.uint8)

fig ,ax = plt.subplots()
ax.plot(t)
ax.set_aspect('equal')
assert len(t) == 256
g = cv.LUT(f,t)

cv.namedWindow('Image',cv.WINDOW_AUTOSIZE)
cv.imshow('Image',f)
cv.waitKey(0)
cv.imshow('Image',g)
cv.waitKey(0)
cv.destroyAllWindows()

h = cv.cvtColor(f,cv.COLOR_BGR2RGB)
i = cv.cvtColor(g,cv.COLOR_BGR2RGB)
fig,ax = plt.subplots(2,1,figsize = (17,10))
ax[0].imshow(h)
ax[1].imshow(i)
plt.show
```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

f = cv.imread(r'C:\Users\Asus\Desktop\Image Processing\zion_pass.jpg')
assert f is not None

hsv = cv.cvtColor(f, cv.COLOR_BGR2HSV)

lower_blue = np.array([95,50,50])
upper_blue = np.array([130,255,255])

lower_green = np.array([26,10,30])
upper_green = np.array([97,100,255])

blue_mask = cv.inRange(hsv,lower_blue,upper_blue)   
green_mask = cv.inRange(hsv,lower_green,upper_green) 

hsv[:,:,0][blue_mask==255] += 10
hsv[:,:,0][green_mask==255] += 50

g = cv.cvtColor(hsv,cv.COLOR_HSV2BGR)
h = cv.cvtColor(f,cv.COLOR_BGR2RGB)
fig,ax = plt.subplots(2,1,figsize = (17,10))
ax[0].imshow(h)
ax[1].imshow(g)
plt.show

cv.namedWindow('Image',cv.WINDOW_AUTOSIZE)
cv.imshow('Image',f)
cv.waitKey(0)
cv.imshow('Image',g)
cv.waitKey(0)
cv.destroyAllWindows()
```
